```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.metrics import classification_report
```

```python
x_train = np.load('x_train.npy')
x_test = np.load('x_test.npy')
y_train = np.load('y_train.npy')
y_test = np.load('y_test.npy')
```

```python
fig,axarr = plt.subplots(2,2,figsize=(15,20))
axarr[0,0].imshow(x_train[0],interpolation='nearest')
axarr[1,0].imshow(x_train[1],interpolation='nearest')
axarr[0,1].imshow(x_train[2],interpolation='nearest')
axarr[1,1].imshow(x_train[3],interpolation='nearest')
```

```python
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
y_train
```

```python
pca=PCA(n_components=120)
x_train=pca.fit_transform(x_train.reshape(2000,128*128))
x_test=pca.fit_transform(x_test.reshape(120,128*128))

```

```python
x_train.shape
```

```python
x_test.shape
```

```python
## KNN
```

```python
clf = KNeighborsClassifier(n_neighbors=1).fit(x_train,y_train)
clf_score = clf.score(x_test,y_test)
print('Accuracy',clf_score)
ypd = clf.predict(x_test)
# cm = confusion_matrix(y_test,ypd,labels=clf.classes_)
# disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=clf.classes_)
cls_rep = classification_report(y_test,ypd)
cls_rep
# disp.plot
```

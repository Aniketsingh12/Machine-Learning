```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df= pd.read_csv('creditcard.csv')
df.head(10)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Class'].value_counts()
```

```python
x=df.drop(columns='Class',axis=1)
x.head()
```

```python
y=df['Class']
y.head()
```

```python
!pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
UtoB = SMOTE(sampling_strategy='auto',random_state=35,k_neighbors=4)
x1,y1 = UtoB.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.33,random_state=33)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(solver = 'liblinear')
```

```python
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
ypd = lr.predict(x_test)
score = accuracy_score(ypd,y_test)
score
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,ypd,labels=lr.classes_)
cm
```

```python
from sklearn.metrics import f1_score
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cmd = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=lr.classes_)
cmd.plot()
```

```python
from sklearn.metrics import recall_score
```

```python
print(recall_score(ypd,y_test,average='binary'))
```

```python
    from sklearn.metrics import precision_score
```

```python
print(precision_score(y_test,ypd,average='binary'))
```

```python
print(f1_score(ypd,y_test,average='binary'))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,ypd))
```
